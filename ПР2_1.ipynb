```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
players=pd.read_csv('/content/players.csv')
pd.set_option('display.max_rows', 5)
pd.set_option('display.max_columns', 16)
players.head(3)
```

```python
players.info()
```

```python
cols = ['FullName', 'Age', 'Height', 'Weight', 'Nationality', 'Club', 'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'ShotPower', 'Stamina', 'Aggression']
players = players[cols]
players.head()
```

```python
fig=plt.figure(figsize=(20,5))
players['Age'].value_counts().plot.bar()
```

```python
fig = plt.figure(figsize = (20,5))
players['Age'].value_counts().sort_index().plot.bar()
```

```python
fig = plt.figure(figsize = (20,5))
(players['Age'].value_counts() / len(players)).plot.bar()
```

```python
players.groupby(['Age'])[['SprintSpeed', 'Stamina', 'ShotPower']].mean()
```

```python
players.groupby(['Age'])[['SprintSpeed', 'Stamina']].mean().plot.bar(figsize = (20, 5), stacked = True)
```

```python
fig = plt.figure(figsize = (20,5))
players['Age'].plot.hist(bins = 10)
```

```python
fig = plt.figure(figsize = (20,5))
players['SprintSpeed'].plot.hist(bins = 2)
```

```python
players[players['Age']<35].sample(200).plot.scatter(x = 'SprintSpeed', y = 'Age', figsize = (20,5))
```

```python
players[players['Age']<35].plot.scatter(x = 'SprintSpeed', y = 'Age', figsize = (20,5))
```

```python
players[players['Age']<35].plot.hexbin(x = 'SprintSpeed', y = 'Age', figsize = (15,7),gridsize = 10, cmap = 'Reds')
```

```python
fig, axes = plt.subplots(2, 3)
fig.set_figheight(5)
fig.set_figwidth(15)

axes[0, 2].scatter(np.arange(30), np.arange(30) + 3 * randn(30))
axes[1, 0].bar(np.arange(10), randn(10))
```

```python
fig, axes = plt.subplots(2, 2, sharex = True, sharey = True)
fig.set_figheight(5)
fig.set_figwidth(15)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(randn(300), bins = 30, color = 'g', alpha = 0.5)
plt.subplots_adjust(wspace = 0.0, hspace = 0.0)
```
