```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
bitc = pd.read_csv("/content/data.csv", index_col=0)
bitc
```

```python
fig = plt.figure(figsize = (20,6))
bpl = plt.plot(bitc)
plt.xticks([])
```

```python
bitc.isnull().sum() #ищем нулевые значения
```

```python
fig = plt.figure(figsize = (20,5))
bpl = plt.plot(bitc)
plt.xticks([])
```

```python
bitc['Bitcoin'].value_counts()
```

```python
bitc[bitc['Bitcoin'] > 80000].value_counts()
```

```python
bitc.loc[bitc['Bitcoin'] > 80000, 'Bitcoin'] = np.nan
bitc = bitc.fillna(method='bfill')
```

```python
fig = plt.figure(figsize = (16,5))
bpl = plt.plot(bitc)
plt.xticks([])
```

```python
bitc[bitc['Bitcoin'] < 0].value_counts()
```

```python
bitc.loc[bitc['Bitcoin'] < 1000, 'Bitcoin'] = np.nan
bitc = bitc.fillna(method='bfill')
```

```python
fig = plt.figure(figsize = (20,5))
bpl = plt.plot(bitc)
plt.xticks([])
```

```python
df = bitc.iloc[107:110]
fig = plt.figure(figsize = (20,5))
bpl = plt.plot(df)
```
